```python
#TASK 5 - Frequency distribution

text1 = "Who doesn't love receiving a little blue box all of their jewelry is clean and classy sleek and just about does anything. One of my favorite necklaces from Tiffany's is one that my best friends got me for graduation when it broke the customer service was awesome with getting it fixed. They have beautiful pieces and although they are lovely , I find them over priced on sterling silver, that is my opinion, we have many of the pieces for my daughter that were given to her as gifts. My husband and I went in to a Tiffany's to shop for an engagement ring and were hardly given the time of day by the sales people.‚Äù al jeweler for a much lower price!"
fd = nltk.FreqDist(text1.split())
fd
```

```python
#TASK 6 - Conditional Frequency distribution
from nltk.probability import ConditionalFreqDist
cfd = ConditionalFreqDist((len(word),word)for word in text1.split())
cfd[5]
```

```python
from wordcloud import WordCloud, STOPWORDS 
import matplotlib.pyplot as plt 
import pandas as pd 
stopwords = set(STOPWORDS)
comment_words = ''
comment_words += " ".join(inauguraltext)+" "
wordcloud = WordCloud(width = 800, height = 800, 
                background_color ='white', 
                stopwords = stopwords, 
                min_font_size = 10).generate(comment_words) 
  
# plot the WordCloud image                        
plt.figure(figsize = (8, 8), facecolor = None) 
plt.imshow(wordcloud) 
plt.axis("off") 
plt.tight_layout(pad = 0) 
  
plt.show() 
  
```
