```python
import matplotlib.pyplot as plt
import string
from langdetect import detect
from sklearn.feature_extraction.text import TfidfVectorizer
from nltk.sentiment.vader import SentimentIntensityAnalyzer
from collections import Counter
from scipy import stats
from dateutil import parser
import seaborn as sns
import shapely
import pandas as pd
df = pd.read_csv('yale.csv')
#df.head()
temp = []
for i in df['Content']:
    i = (i.replace('<p>','')).strip('][')
    i = i.replace("</p>","")
    i = i.replace("</a>","")
    i = i.replace("/n","")
    temp.append(i.replace("<a>",""))
temp
df['Content'] = temp
```

```python
vectorizer = TfidfVectorizer(min_df=5, max_features=10, stop_words='english', ngram_range=(1,1), use_idf=True)
matrix = vectorizer.fit_transform(df['Content'])
```

```python
import nltk
analyzer = SentimentIntensityAnalyzer()
def sentiment_analyzer_scores(comment):
    score = analyzer.polarity_scores(comment)
    return(pd.DataFrame.from_dict(score, orient='index'))
sentiment_analyzer_scores(df['Content'][0])
```

```python
polarity_compound = lambda s: (analyzer.polarity_scores(s))['compound']
```

```python
df['polarity'] = df['Content'].map(polarity_compound)
```

```python
df[:5].style.bar(subset=['polarity'], align='mid', color=['#d65f5f', '#5fba7d'])
```

```python
df[['Title','polarity']]
```
