```python
import nltk
```

```python
sentence = [("the", "DT"), ("little", "JJ"), ("yellow", "JJ"),("dog", "NN"), ("barked", "VBD"), ("at", "IN"),  ("the", "DT"), ("cat", "NN")]
grammar = "NP: {<DT>?<JJ>*<NN>}"
cp = nltk.RegexpParser(grammar)
result = cp.parse(sentence)
```

```python
print(result)
```

```python
result.draw()
```

```python
f = open("Gettysburg Address.txt","r")
content = f.read()
print(content)
f.close()
```

```python
#Sentence Tokenize
from nltk.tokenize import sent_tokenize 
sen_token = sent_tokenize(content)
sen_token
```

```python
#Word Tokenize
from nltk.tokenize import word_tokenize 
word_token = word_tokenize(content)
word_token
```

```python
pos = nltk.pos_tag(word_token)
pos
```

```python
sentence = pos
grammar = "VP: {<MD>?<VB.*><NP|PP>}" 
cp = nltk.RegexpParser(grammar)
result = cp.parse(sentence)
print(result)
```

```python
result.draw()
```
